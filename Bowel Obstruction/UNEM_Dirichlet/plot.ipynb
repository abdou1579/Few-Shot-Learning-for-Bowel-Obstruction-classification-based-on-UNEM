```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

datasets = ['oxfordpets', 'eurosat', 'ucf101', 'caltech101', 'dtd', 'fgvcaircraft', 'food101', 'flowers102', 'stanfordcars', 'sun397']
t = []
gamma = []
label = []
labels = ['OxfordPets', 'EuroSAT', 'UCF101', 'Caltech101', 'DTD', 'FGVC Aircraft', 'Food101', 'Flowers102', 'Stanford Cars', 'SUN397']
plt.figure(dpi=1200)
plt.rcParams['font.family'] = 'Times New Roman'
for dataset in datasets:
    n = 8 if dataset == 'sun397' else 10
    p = torch.load(f'u_params/{dataset}/RN50/{n}/U_EM_DIRICHLET_EMBT_L_4shots.pt')
    t1_temp = []
    t2_temp = []
    gamma_temp = []
    print(p)
    for key in p.keys():
        if key.startswith('emb_t'):
            t1_temp.append(nn.Softplus()(p[key]).item())
        elif key.startswith('t'):
            t2_temp.append(nn.Softplus()(p[key]).item()+1)
        elif key.startswith('gamma'):
            gamma_temp.append(nn.Softplus()(p[key]).item())
        else:
            raise ValueError(f"Unknown key: {key}")
    avg_t1 = sum(t1_temp)/len(t1_temp)
    avg_t2 = sum(t2_temp)/len(t2_temp)
    avg_gamma = sum(gamma_temp)/len(gamma_temp)
    t.append(avg_t2)
    gamma.append(avg_gamma)
    label.append(dataset)


for i, txt in enumerate(labels):
    plt.scatter(gamma[i]*75, t[i], marker='h', s=100, label=txt)
#     text= plt.text(row['lambda']*75+0.5, softplus(row['t2'])+1.005, f"{row['dataset']}", fontsize=9)
#     texts.append(text)
# adjust_text(texts)

plt.xlabel(r"$\lambda$", fontsize='large')
plt.ylabel("T", fontsize='large')
plt.grid(linestyle=':')
plt.legend(fontsize='large', loc='lower right')
plt.savefig('plot/plot_avg_params_EM.png')
```

```python
import torch
import torch.nn as nn

datasets = ['caltech101', 'dtd', 'eurosat', 'fgvcaircraft', 'flowers102', 'food101', 'oxfordpets', 'stanfordcars', 'ucf101']
y_labels = []
x_labels = []
for dataset in datasets:
    p = torch.load(f'u_params/{dataset}/RN50/10/U_EM_DIRICHLET_EMBT_L_4shots.pt')
    x_labels_temp = []
    y_labels_temp = []
    for key in p.keys():
        x_labels_temp.append(key)
        if key.startswith('emb_t'):
            y_labels_temp.append(nn.Softplus()(p[key]).item())
        elif key.startswith('t'):
            y_labels_temp.append(nn.Softplus()(p[key]).item()+1)
        elif key.startswith('gamma'):
            y_labels_temp.append(nn.Softplus()(p[key]).item())
        else:
            raise ValueError(f"Unknown key: {key}")
    x_labels.append(x_labels_temp)
    y_labels.append(y_labels_temp)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(3, 3, figsize=(12, 10), dpi=600)
# flatten the axis array
ax = ax.flatten()
for i in range(9):
    ax[i].scatter(x=x_labels[i], y=y_labels[i], color='blue', marker='o')  # Use range(len(y_labels)) for equally spaced points
    ax[i].set_xticks(range(len(x_labels[i])))
    ax[i].set_xticklabels(x_labels[i], fontsize=8, rotation=70)
    ax[i].set_ylabel("Values")
    ax[i].set_title(datasets[i])

plt.tight_layout()
# Display the plot
plt.show()
    
```
